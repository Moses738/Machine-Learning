```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
data=sns.load_dataset('tips')
```

```python
data.head()
```

## Distribution

```python
sns.displot(data)
plt.title("Distribution of Tips data")
plt.show()
```

```python
sns.displot(data,kind='kde')
plt.title("Distribution of Tips data")
plt.show()
```

## Scaterplot

```python
sns.scatterplot(x=data.total_bill,y=data.tip)
plt.title("The Relation between total_bill vs Tips")
plt.show()
```

> Where the total_bill is increasing tip also increasing\
> They are positively correlated

## Joint Plot

```python
data.columns
```

```python
plt.figure(figsize=(7,10))
sns.jointplot(x=data.total_bill,y=data.size)
plt.title("The total_bill vs size")
plt.show()
```

## Boxplot

```python
data.columns
```

```python
sns.boxplot(x=data.sex,y=data.total_bill) #whiskers,25%,50%,75%,outliers
plt.show()
```

> In male 25% to 75% of total bills are distributed in between 15 rupees to 25rupees\
> In male we are having some outliers\
> In female also we are having some outliers\
> Male higher in total bills and female is lower than male

## Barplot

```python
sns.barplot(x=data.smoker,y=data.total_bill) # Confidence interval
plt.show()
```

## Countplot

```python
sns.countplot(data.smoker) # Confidence interval
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.countplot(data.day) # Confidence interval
plt.show()
```

```python
count=sns.countplot(data.day)
for i in count.containers:
    count.bar_label(i)
plt.title("The count of Day")
plt.show()
```

## Pair Plot

```python
sns.pairplot(data)
plt.show()
```
